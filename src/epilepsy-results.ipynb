# Epilepsy and Tumor Diagnosis Results

GDNN1, DNN1, DNN3 and GDNN3 loss and accuracy

```python
import numpy as np
import os
import math

path = os.getcwd()
path 
epiNN5 = np.loadtxt(path + "/../results/epilepsy-5NN_out.csv", delimiter=",")
epiNN3 = np.loadtxt(path + "/../results/epilepsy-3NN_out.csv", delimiter=",")
```

```python
from matplotlib.pyplot import figure
import matplotlib.pyplot as plt
%matplotlib inline

figure(num=None, figsize=(8, 4), dpi=80, facecolor='w', edgecolor='k')

plt.plot(epiNN3[:,0],epiNN3[:,1], label="epilepsy-3NN")
plt.fill_between(epiNN3[:,0], epiNN3[:,1]+epiNN3[:,1]*epiNN3[:,2],  epiNN3[:,1]-epiNN3[:,1]*epiNN3[:,2], alpha=0.4)

plt.plot(epiNN5[:,0],epiNN5[:,1], label="epilepsy-5NN")
plt.fill_between(epiNN5[:,0], epiNN5[:,1]+epiNN5[:,1]*epiNN5[:,2],  epiNN5[:,1]-epiNN5[:,1]*epiNN5[:,2], alpha=0.4)

plt.grid(True)
plt.title('Epilepsy training process')
plt.xlabel('epochs')
plt.ylabel('loss');
plt.legend()
plt.show()
```

```python
from matplotlib.pyplot import figure
import matplotlib.pyplot as plt
%matplotlib inline

figure(num=None, figsize=(8, 4), dpi=80, facecolor='w', edgecolor='k')

plt.plot(epiNN3[:,0],epiNN3[:,3], label="epilepsy-3NN")
plt.fill_between(epiNN3[:,0], epiNN3[:,3]+epiNN3[:,3]*epiNN3[:,4]*0.1,  epiNN3[:,3]-epiNN3[:,3]*epiNN3[:,4], alpha=0.2)

plt.plot(epiNN5[:,0],epiNN5[:,3], label="epilepsy-5NN")
plt.fill_between(epiNN5[:,0], epiNN5[:,3]+epiNN5[:,3]*epiNN5[:,4]*0.1,  epiNN5[:,3]-epiNN5[:,3]*epiNN5[:,4], alpha=0.4)

plt.grid(True)
plt.title('Epilepsy perfomance on test partition ')
plt.xlabel('epochs')
plt.ylabel('accuracy');
plt.legend()
plt.show()
```
