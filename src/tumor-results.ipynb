# Epilepsy and Tumor Diagnosis Results

GDNN1, DNN1, DNN3 and GDNN3 loss and accuracy

```python
import numpy as np
import os
import math

path = os.getcwd()
path 
tumNN5 = np.loadtxt(path + "/../results/tumor-5NN_out.csv", delimiter=",")
```

```python
from matplotlib.pyplot import figure
import matplotlib.pyplot as plt
%matplotlib inline

figure(num=None, figsize=(8, 4), dpi=80, facecolor='w', edgecolor='k')

plt.plot(tumNN5[:,0],tumNN5[:,1], label="epilepsy-5NN")
plt.fill_between(tumNN5[:,0], tumNN5[:,1]+tumNN5[:,1]*tumNN5[:,2],  tumNN5[:,1]-tumNN5[:,1]*tumNN5[:,2], alpha=0.4)

plt.grid(True)
plt.title('Epilepsy training process')
plt.xlabel('epochs')
plt.ylabel('loss');
plt.legend()
plt.show()
```
