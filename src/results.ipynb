# Epilepsy and Tumor Diagnosis Results

GDNN1, DNN1, DNN3 and GDNN3 loss and accuracy

```python
import numpy as np
import os
import math

path = os.getcwd()
path 
epiNN5 = np.loadtxt(path + "/../results/epilepsy-5NN.out", delimiter=",")

# dnn1 = np.loadtxt(path + "/../results/DNN1.csv", delimiter=",")
# gdnn3 = np.loadtxt(path + "/../results/GDNN3.csv", delimiter=",")
# dnn3 = np.loadtxt(path + "/../results/DNN3.csv", delimiter=",")

epiNN5 = epiNN5[:100,:]
```

```python
from matplotlib.pyplot import figure
import matplotlib.pyplot as plt
%matplotlib inline

figure(num=None, figsize=(8, 4), dpi=80, facecolor='w', edgecolor='k')

plt.plot(epiNN5[:,0],epiNN5[:,1], label="epilepsy")
plt.fill_between(epiNN5[:,0], epiNN5[:,1]+epiNN5[:,1]*epiNN5[:,2],  epiNN5[:,1]-epiNN5[:,1]*epiNN5[:,2], alpha=0.4)

# plt.gca().set_xscale("log", nonposx='clip')
# plt.gca().set_yscale("log", nonposy='clip')
plt.grid(True)
plt.title('Epilepsy and Tumor training process')
plt.xlabel('epochs')
plt.ylabel('loss');
plt.legend()
plt.show()
```
